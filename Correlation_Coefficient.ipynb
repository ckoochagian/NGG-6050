```python
from matplotlib import pyplot as plt
import numpy as np

#raw data
x = [10.4, 10.8, 11.1, 10.2, 10.3, 10.2, 10.7, 10.5, 10.8, 11.2, 10.6, 11.4] #wing length
y = [7.4, 7.6, 7.9, 7.2, 7.4, 7.1, 7.4, 7.2, 7.8, 7.7, 7.8, 8.8] #tail length

#visualizing the data
plt.scatter(x, y)

#making graph pretty
plt.title("correlation between wing length and tail length")
plt.xlabel("wing length")
plt.ylabel("tail length")

plt.show()

#calculate r_xy and r_yx using math
n = 12
x_bar = np.sum(x) / n
y_bar = np.sum(y) / n
numerator = sum((xi - x_bar) * (yi - y_bar) for xi, yi in zip (x, y))
denom_x = np.sqrt(sum([(xi - x_bar)**2 for xi in x]))
denom_y = np.sqrt(sum([(yi - y_bar)**2 for yi in y]))
r = numerator / (denom_x * denom_y)
print(f"Correlation coefficient using math: {r:.2f}")

#calculate r_xy anf r_yx using python
correlation_matrix = np.corrcoef(x, y)
r_xy = correlation_matrix[0, 1]
r_yx = correlation_matrix[1, 0]
print(f"Correlation coefficient for xy from python: {r_xy:.2f} ; Correlation coefficient for yx from python: {r_yx:.2f}")
```
