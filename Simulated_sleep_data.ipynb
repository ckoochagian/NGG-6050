```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Parameters
num_days = 1  # Simulate sleep data for 1 day
bins_per_day = 48  # 30-minute bins, so 24 hours * 2 = 48 bins/day
total_bins = num_days * bins_per_day

# Simulate sleep patterns
# Daytime: Flies are generally awake (e.g., bins 16 to 32 might represent day).
# Nighttime: Flies are generally asleep (e.g., bins 0-15, 33-47 represent night).

# Probability of sleeping in each time bin
sleep_prob_day = 0.2
sleep_prob_night = 0.9

# Create an array for sleep probability based on circadian pattern (night vs day)
sleep_prob = np.array([sleep_prob_night]*16 + [sleep_prob_day]*16 + [sleep_prob_night]*16)

# Repeat the pattern for multiple days
sleep_prob_full = np.tile(sleep_prob, num_days)

# Simulate sleep (1 for sleep, 0 for awake)
sleep_data = np.random.binomial(n=30, p=sleep_prob_full)

# Create a timestamp index for 30-minute intervals
time_index = pd.date_range(start='2024-01-01', periods=total_bins, freq='30min')

# Create a DataFrame to hold the sleep data
df_sleep = pd.DataFrame({'Sleep': sleep_data}, index=time_index)


#now we need to simulate sleep pattern if immortalized neurons promote sleep upon activation
# would use thermogenetics during day, easier to see if there is a sleep promoting effect when flies are generally active
# Daytime: Flies are generally awake (e.g., bins 16 to 32 might represent day).
# Nighttime: Flies are generally asleep (e.g., bins 0-15, 33-47 represent night).

# Probability of sleeping in each time bin
sleep_prob_day_immortal = 0.75 #arbitrary probability, but we would predict increased probability during day if activating immortalized neurons promote sleep in adulthood
sleep_prob_night_immortal = 0.9

# Create an array for sleep probability based on circadian pattern (night vs day)
sleep_prob_immortal = np.array([sleep_prob_night_immortal]*16 + [sleep_prob_day_immortal]*16 + [sleep_prob_night_immortal]*16)

# Repeat the pattern for multiple days
sleep_prob_full_immortal = np.tile(sleep_prob_immortal, num_days)

# Simulate sleep (1 for sleep, 0 for awake)
sleep_data_immortal = np.random.binomial(n=30, p=sleep_prob_full_immortal)

# Create a timestamp index for 30-minute intervals
time_index = pd.date_range(start='2024-01-01', periods=total_bins, freq='30min')

# Create a DataFrame to hold the sleep data
df_sleep_immortal = pd.DataFrame({'Sleep': sleep_data_immortal}, index=time_index)





# Plotting the sleep pattern
plt.figure(figsize=(12, 4))
plt.plot(df_sleep.index, df_sleep['Sleep'], drawstyle='steps-pre', label='Control', color='blue')
plt.plot(df_sleep_immortal.index, df_sleep_immortal['Sleep'], drawstyle='steps-pre', label='Immortalized R76G11', color='red')
plt.title('Simulated Data for Thermogenetic Activation of Immortalized R76G11 Neurons in Adulthood')
plt.xlabel('Time (hrs)')
plt.ylabel('Sleep (mins/30 min)')
plt.legend()
plt.show()



```
